```python
import pandas as pd
import matplotlib as plt
from geopy.geocoders import Nominatim
```

```python
file_path = 'Superstore.xls'
df = pd.read_excel(file_path, sheet_name='Orders')
df.head()
```

```python
import time

#create a function to get latitude and longitude information using geopy
def get_lat_lon(address):
    app = Nominatim(user_agent="test")
    #location = app.geocode(address, timeout=3)
    #location = app.geocode(address)
    location = app.geocode(address, timeout=None)
    return location.latitude, location.longitude
```

```python
#combine 'City' and 'State' info to create 'Address' column info
#df['Address'] = df['City'] + " " + df["State"]
df['Address'] = df['City'] + ", " + df["State"]

address_list = df['Address'].unique()
address_list = address_list.tolist()
lat_list = []
lon_list = []
i = 0

for a in address_list:
    #to prevent the timeout error, take 30 sec time.sleep every 40 addresses
    if i > 40:
        time.sleep(30)
        i = 0
    
    lat, lon = get_lat_lon(a)
    lat_list.append(lat)
    lon_list.append(lon)
    i += 1
    print(i)
```

```python
#check total numbers of lat_list, lon_list and address_list
print(len(lat_list), len(lon_list), len(address_list))
```

```python
#create a pandas DataFrame to store the info of lat_list, lon_list and address_list
lat_lon_df = pd.DataFrame({'address': [], 'lat': [], 'lon': []})
lat_lon_df['address'] = address_list
lat_lon_df['lat'] = lat_list
lat_lon_df['lon'] = lon_list
lat_lon_df
```

```python
#export the lat_lon_df to a csv file
lat_lon_df.to_csv('lat_lon.csv', index=False)
```
