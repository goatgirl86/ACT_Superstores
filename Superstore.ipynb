```python
df_c = df.groupby(['Category'], as_index=False)['Profit'].sum()
df_c
```

```python
df_s = df.groupby(['State'], as_index=False)['Profit'].mean().sort_values(['Profit'],ascending=False)
# df_s=df_s.set_index(['State'])
df_s.head(5)
```

```python
state_mean=df.groupby("State")["Profit"].agg([np.mean, np.sum]).sort_values(['sum'], ascending=False)
state_mean.head(5)
```

```python
state_mean=df.groupby("State")["Sales"].agg([np.mean, np.sum]).sort_values(['mean'], ascending=False)
state_mean.head(5)
```

```python
df_scsr = df.groupby(['State','Region'], as_index=False)['Profit'].mean()
df_scsr.head(5)
```

```python
df_scsr['Percent Of Region Sales'] = df_scsr['Profit'] / df_scsr.groupby('Region')['Profit'].transform('sum')
df_scsr.head()
```

```python
state_mean2=df.groupby(["State", "Region"])["Profit"].agg([np.mean, np.sum]).sort_values(['mean'], ascending=False)
state_mean2.head(5)
```

```python
state_mean2['Percent Of Region Mean Sales'] = state_mean2['mean'] / state_mean2.groupby('Region')['mean'].transform('sum') * 100
state_mean2.head(5)
```

```python
state_mean2['Percent Of Region Sum Sales'] = state_mean2['sum'] / state_mean2.groupby('Region')['sum'].transform('sum') * 100
state_mean2.head(5)
```

```python
state_mean2=state_mean2.sort_values(['Region','mean'], ascending=False)
state_mean2.head(5)
```
